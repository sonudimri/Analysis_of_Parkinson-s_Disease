## outlier detection

```python
# Get influence object
influence = OLSInfluence(model)

# -------------------------------
# 1. Cook's Distance
# -------------------------------
cooks_d = influence.cooks_distance[0]
threshold_cook = 4 / len(X)

plt.figure(figsize=(10, 4))
plt.stem(cooks_d, markerfmt=",")
plt.axhline(y=threshold_cook, color='red', linestyle='--', label=f'Threshold: {round(threshold_cook, 4)}')
plt.title("Cook's Distance")
plt.xlabel("Observation Index")
plt.ylabel("Cook's Distance")
plt.legend()
plt.show()

# -------------------------------
# 2. DFFITS
# -------------------------------
dffits, _ = influence.dffits
threshold_dffits = 2 * np.sqrt(X.shape[1] / len(X))

plt.figure(figsize=(10, 4))
plt.stem(dffits, markerfmt=",")
plt.axhline(y=threshold_dffits, color='red', linestyle='--', label=f'Threshold: {round(threshold_dffits, 4)}')
plt.axhline(y=-threshold_dffits, color='red', linestyle='--')
plt.title("DFFITS")
plt.xlabel("Observation Index")
plt.ylabel("DFFITS")
plt.legend()
plt.show()

# -------------------------------
# 3. DFBETAs (for all predictors)
# -------------------------------
dfbetas = influence.dfbetas
threshold_dfbeta = 2 / np.sqrt(len(X))

# Optional: Check max DFBETA per obs
dfbeta_max = np.max(np.abs(dfbetas), axis=1)

plt.figure(figsize=(10, 4))
plt.stem(dfbeta_max, markerfmt=",")
plt.axhline(y=threshold_dfbeta, color='red', linestyle='--', label=f'Threshold: {round(threshold_dfbeta, 4)}')
plt.title("Max DFBETA per Observation")
plt.xlabel("Observation Index")
plt.ylabel("Max DFBETA")
plt.legend()
plt.show()

# -------------------------------
# 4. COVRATIO
# -------------------------------
cov_ratio = influence.cov_ratio
p = X.shape[1]
n = X.shape[0]
upper = 1 + 3 * p / n
lower = 1 - 3 * p / n

plt.figure(figsize=(10, 4))
plt.plot(cov_ratio, 'o')
plt.axhline(y=upper, color='red', linestyle='--', label=f'Upper: {round(upper, 3)}')
plt.axhline(y=lower, color='red', linestyle='--', label=f'Lower: {round(lower, 3)}')
plt.title("COVRATIO")
plt.xlabel("Observation Index")
plt.ylabel("COVRATIO")
plt.legend()
plt.show()

# -------------------------------
# Flagging potential outliers
# -------------------------------
outliers = pd.DataFrame({
    'Cook': cooks_d > threshold_cook,
    'DFFITS': np.abs(dffits) > threshold_dffits,
    'Max_DFBETA': dfbeta_max > threshold_dfbeta,
    'COVRATIO': (cov_ratio > upper) | (cov_ratio < lower)
})

# How many were flagged
print("Outliers flagged by each method:")
print(outliers.sum())

# Get all rows flagged by any method
suspect_obs = outliers.any(axis=1)
print(f"\nTotal suspect observations: {suspect_obs.sum()}")

# View them
print(df[suspect_obs])
```

## 2. Condition Indices and Variance Decomposition Proportions

```python
import statsmodels.api as sm

# Make sure X includes the constant column (and is already standardized except const)
X_values = X.values
X_names = X.columns

# Step 1: Singular Value Decomposition (SVD)
U, s, Vt = np.linalg.svd(X_values, full_matrices=False)

# Step 2: Condition Indices
condition_indices = s[0] / s
condition_indices = np.round(condition_indices, 2)

# Step 3: Eigenvalues and Inverses
eigvals = s**2
inv_eigvals = 1 / eigvals

# Step 4: Variance Decomposition Proportions (VDP)
phi = (Vt.T ** 2) * inv_eigvals.reshape(1, -1)
phi = phi / phi.sum(axis=1, keepdims=True)

# Step 5: Assemble into DataFrame
phi_df = pd.DataFrame(phi, columns=[f"CI_{i+1}" for i in range(len(s))])
phi_df["feature"] = X_names
phi_df = phi_df.set_index("feature").T

# Add Condition Indices
phi_df.insert(0, "Condition Index", condition_indices)

# Display full result
pd.set_option('display.max_columns', None)
print(phi_df)
```

##  yeo-johnson 

```python
from sklearn.preprocessing import PowerTransformer
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Define predictors and target
X = df1.drop(columns=['motor_UPDRS', 'total_UPDRS'])
Y = df1['motor_UPDRS']  # or 'total_UPDRS'

# 2. Apply Yeo-Johnson transformation to X
pt = PowerTransformer(method='yeo-johnson')
transformed_X = pt.fit_transform(X)

# 3. Add constant for intercept
X_trans = sm.add_constant(transformed_X)

# 4. Fit OLS model
model = sm.OLS(Y, X_trans).fit()
print(model.summary())

# 5. Residuals and fitted values
residuals = model.resid
fitted = model.fittedvalues

# --- Q-Q Plot of Residuals ---
sm.qqplot(residuals, line='45', fit=True)
plt.title("Q-Q Plot of Residuals (Yeo-Johnson Transformed)")
plt.grid(True)
plt.show()

# --- Actual vs Fitted Plot ---
plt.figure(figsize=(8, 5))
sns.scatterplot(x=fitted, y=Y, alpha=0.6)
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], color='red', linestyle='--')
plt.xlabel("Fitted Values")
plt.ylabel("Actual Values")
plt.title("Actual vs Fitted (Yeo-Johnson Transformed)")
plt.grid(True)
plt.show()
```

## lets try to solve using Ridge and lasso

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV, LassoCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer

# Apply Yeo-Johnson transformation to features and target
pt_x = PowerTransformer(method='yeo-johnson')
X_transformed = pt_x.fit_transform(X)

pt_y = PowerTransformer(method='yeo-johnson')
y_transformed = pt_y.fit_transform(Y.values.reshape(-1, 1)).flatten()

# Split the transformed data
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y_transformed, test_size=0.2, random_state=42)

# Ridge Regression
ridge = RidgeCV(alphas=np.logspace(-3, 3, 100), cv=5)
ridge.fit(X_train, y_train)
ridge_r2 = ridge.score(X_test, y_test)

# Lasso Regression
lasso = LassoCV(alphas=np.logspace(-3, 3, 100), cv=5, max_iter=10000)
lasso.fit(X_train, y_train)
lasso_r2 = lasso.score(X_test, y_test)

# Print performance metrics
print(f"Ridge R² (transformed data): {ridge_r2:.4f} | Best alpha: {ridge.alpha_:.4f}")
print(f"Lasso R² (transformed data): {lasso_r2:.4f} | Best alpha: {lasso.alpha_:.4f}")

# Function to plot coefficients with original feature names
def plot_transformed_coefficients(model, feature_names, model_name, color):
    """Plot model coefficients with proper feature alignment"""
    try:
        coef = model.coef_
        if coef.ndim > 1:
            coef = coef.ravel()
            
        if len(coef) != len(feature_names):
            raise ValueError(f"Feature count mismatch: {len(coef)} coefficients vs {len(feature_names)} features")
            
        coef_series = pd.Series(coef, index=feature_names)
        
        plt.figure(figsize=(10,6))
        coef_series.sort_values().plot(kind='barh', color=color)
        plt.title(f"{model_name} Coefficients on Transformed Data (α={model.alpha_:.4f})")
        plt.xlabel("Coefficient Value")
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"Error plotting {model_name} coefficients: {str(e)}")

# Plot coefficients using original feature names
plot_transformed_coefficients(ridge, X.columns, "Ridge Regression", 'steelblue')
plot_transformed_coefficients(lasso, X.columns, "Lasso Regression", 'coral')

# Show features selected by Lasso
print("\nFeatures selected by Lasso (non-zero coefficients) on transformed data:")
print(X.columns[lasso.coef_ != 0].tolist())

# Optional: Compare performance with/without transformation
print("\nNote: These coefficients are for the transformed features. For interpretation,")
print("you may need to inverse transform the predictions to original scale.")
```
