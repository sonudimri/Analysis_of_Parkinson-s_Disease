## Regression Modelling Part 

```python
# Add constant for intercept
X = sm.add_constant(scaled_X)

# Fit model
model = sm.OLS(Y, X).fit()
model.summary()
```

```python
# Get influence object
influence = OLSInfluence(model)

# -------------------------------
# 1. Cook's Distance
# -------------------------------
cooks_d = influence.cooks_distance[0]
threshold_cook = 4 / len(X)

plt.figure(figsize=(10, 4))
plt.stem(cooks_d, markerfmt=",")
plt.axhline(y=threshold_cook, color='red', linestyle='--', label=f'Threshold: {round(threshold_cook, 4)}')
plt.title("Cook's Distance")
plt.xlabel("Observation Index")
plt.ylabel("Cook's Distance")
plt.legend()
plt.show()

# -------------------------------
# 2. DFFITS
# -------------------------------
dffits, _ = influence.dffits
threshold_dffits = 2 * np.sqrt(X.shape[1] / len(X))

plt.figure(figsize=(10, 4))
plt.stem(dffits, markerfmt=",")
plt.axhline(y=threshold_dffits, color='red', linestyle='--', label=f'Threshold: {round(threshold_dffits, 4)}')
plt.axhline(y=-threshold_dffits, color='red', linestyle='--')
plt.title("DFFITS")
plt.xlabel("Observation Index")
plt.ylabel("DFFITS")
plt.legend()
plt.show()

# -------------------------------
# 3. DFBETAs (for all predictors)
# -------------------------------
dfbetas = influence.dfbetas
threshold_dfbeta = 2 / np.sqrt(len(X))

# Optional: Check max DFBETA per obs
dfbeta_max = np.max(np.abs(dfbetas), axis=1)

plt.figure(figsize=(10, 4))
plt.stem(dfbeta_max, markerfmt=",")
plt.axhline(y=threshold_dfbeta, color='red', linestyle='--', label=f'Threshold: {round(threshold_dfbeta, 4)}')
plt.title("Max DFBETA per Observation")
plt.xlabel("Observation Index")
plt.ylabel("Max DFBETA")
plt.legend()
plt.show()

# -------------------------------
# 4. COVRATIO
# -------------------------------
cov_ratio = influence.cov_ratio
p = X.shape[1]
n = X.shape[0]
upper = 1 + 3 * p / n
lower = 1 - 3 * p / n

plt.figure(figsize=(10, 4))
plt.plot(cov_ratio, 'o')
plt.axhline(y=upper, color='red', linestyle='--', label=f'Upper: {round(upper, 3)}')
plt.axhline(y=lower, color='red', linestyle='--', label=f'Lower: {round(lower, 3)}')
plt.title("COVRATIO")
plt.xlabel("Observation Index")
plt.ylabel("COVRATIO")
plt.legend()
plt.show()

# -------------------------------
# Flagging potential outliers
# -------------------------------
outliers = pd.DataFrame({
    'Cook': cooks_d > threshold_cook,
    'DFFITS': np.abs(dffits) > threshold_dffits,
    'Max_DFBETA': dfbeta_max > threshold_dfbeta,
    'COVRATIO': (cov_ratio > upper) | (cov_ratio < lower)
})

# How many were flagged
print("Outliers flagged by each method:")
print(outliers.sum())

# Get all rows flagged by any method
suspect_obs = outliers.any(axis=1)
print(f"\nTotal suspect observations: {suspect_obs.sum()}")

# View them
print(df[suspect_obs])
```

#### Let's handle Multicollinearity 

##### 1. Variance Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Ensure constant is included in X_sc
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = np.round([variance_inflation_factor(X.values, i) for i in range(X.shape[1])],2)

print(vif_data.sort_values(by="VIF", ascending=False))
```

##### 2. Condition Indices and Variance Decomposition Proportions

```python
import statsmodels.api as sm

# Make sure X includes the constant column (and is already standardized except const)
X_values = X.values
X_names = X.columns

# Step 1: Singular Value Decomposition (SVD)
U, s, Vt = np.linalg.svd(X_values, full_matrices=False)

# Step 2: Condition Indices
condition_indices = s[0] / s
condition_indices = np.round(condition_indices, 2)

# Step 3: Eigenvalues and Inverses
eigvals = s**2
inv_eigvals = 1 / eigvals

# Step 4: Variance Decomposition Proportions (VDP)
phi = (Vt.T ** 2) * inv_eigvals.reshape(1, -1)
phi = phi / phi.sum(axis=1, keepdims=True)

# Step 5: Assemble into DataFrame
phi_df = pd.DataFrame(phi, columns=[f"CI_{i+1}" for i in range(len(s))])
phi_df["feature"] = X_names
phi_df = phi_df.set_index("feature").T

# Add Condition Indices
phi_df.insert(0, "Condition Index", condition_indices)

# Display full result
pd.set_option('display.max_columns', None)
print(phi_df)
```

## Ridge & Lasso Regression

```python
X_reg = scaled_X
y = Y
```

```python
from sklearn.linear_model import RidgeCV, LassoCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(X_reg,y,test_size=0.2)
```

```python
ridge = RidgeCV(alphas=np.logspace(-3, 3, 100), cv=5)
model1 = ridge.fit(X_train, y_train)
ridge_r2 = ridge.score(X_test, y_test)


lasso = LassoCV(alphas=np.logspace(-3, 3, 100), cv=5, max_iter=10000)
model2 = lasso.fit(X_train, y_train)
lasso_r2 = lasso.score(X_test, y_test)

print(f" Ridge R²: {ridge_r2:.4f} | Best alpha: {ridge.alpha_}")
print(f" Lasso R²: {lasso_r2:.4f} | Best alpha: {lasso.alpha_}")
```

```python
ridge_coef = pd.Series(ridge.coef_, index=X_reg.columns)
lasso_coef = pd.Series(lasso.coef_, index=X_reg.columns)

plt.figure(figsize=(10,6))
ridge_coef.sort_values().plot(kind='barh', color='steelblue', label='Ridge')
plt.title("Ridge Coefficients")
plt.show()

plt.figure(figsize=(10,6))
lasso_coef.sort_values().plot(kind='barh', color='coral', label='Lasso')
plt.title("Lasso Coefficients")
plt.show()
```

#### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

ridge_poly = RidgeCV(alphas=np.logspace(-3, 3, 100), cv=5)
ridge_poly.fit(X_train_poly, y_train)
ridge_poly_r2 = ridge_poly.score(X_test_poly, y_test)

print(f" Ridge with poly R²: {ridge_poly_r2:.4f}")
```

## Random Forest

```python
# Train a Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict
y_pred = rf_model.predict(X_test)

# Step 5: Evaluate Model Performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R² Score: {r2:.4f}")

# Plot Actual vs Predicted
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.xlabel("Actual motor_UPDRS")
plt.ylabel("Predicted motor_UPDRS")
plt.title("Actual vs. Predicted motor_UPDRS")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "r--")
plt.show()
```

### K-Fold Validation

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import make_scorer, mean_squared_error, r2_score

# Define K-Fold Cross-Validation (K=5)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42,n_jobs=-1,verbose=1)

# Define scoring metrics
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)
r2_scorer = make_scorer(r2_score)

# Perform Cross-Validation
mse_scores = cross_val_score(rf_model, X, y, cv=kf, scoring=mse_scorer)
r2_scores = cross_val_score(rf_model, X, y, cv=kf, scoring=r2_scorer)

# Compute average scores
avg_mse = -np.mean(mse_scores)  # Convert negative MSE back to positive
avg_r2 = np.mean(r2_scores)

print(f"Cross-Validated Mean Squared Error (MSE): {avg_mse:.4f}")
print(f"Cross-Validated R² Score: {avg_r2:.4f}")
```
